```python
import pandas as pd
df = pd.read_excel('data.xlsx')
df.head()
```

```python
df['tan'] = df['sigma']/(df['w']*df['e'])
def material(df):
  if df['tan'] == 0:
    return 'Perfecto'
  elif df['tan'] <= 25:
    return 'Perdidas'
  elif df['tan'] > 25:
    return 'Conductor'

df['material'] = df.apply(material, axis=1)
df.drop('tan', axis=1, inplace=True)
df.head()
```

```python
data_buen_conductor = df[df['material'] == 'Conductor']
data_buen_conductor.head(14)
```

```python
data_buen_conductor['u'].values
```

```python
minimo = 1.2566370614359173e-06
maximo = 2.5132741228718343e-05
media_objetivo = (maximo + minimo) / 2

# La desviación estándar de una distribución uniforme es sqrt((max - min)^2 / 12)
desviacion_objetivo = ((maximo - minimo) ** 2 / 12) ** 0.5
media_objetivo, desviacion_objetivo
```

```python
import numpy as np
import scipy.stats as stats

# Datos iniciales
datos = np.array([1.25663706e-06, 1.25663706e-06, 1.25663706e-06, 1.25663706e-06,
       1.25663706e-06, 2.51327412e-05, 2.63893783e-06, 6.28318531e-06,
       1.25663706e-06, 1.25663706e-06, 1.25663706e-06, 1.25663706e-06,
       1.25663706e-06, 1.25663706e-06])

np.random.seed(0)  # Para reproducibilidad
contador = len(datos)  # Contador inicial

while contador < 200:
    # Añadir un nuevo dato aleatorio
    nuevo_dato = np.random.normal(media_objetivo, desviacion_objetivo)
    if nuevo_dato >= 0:
      datos = np.append(datos, nuevo_dato)

      # Calcular la nueva media y desviación estándar
      media_actual = np.mean(datos)
      desviacion_actual = np.std(datos)
      contador += 1
      # Verificar si se cumplen los criterios de la media y la desviación estándar
      if abs(media_actual - media_objetivo) < 1e-7 and abs(desviacion_actual - desviacion_objetivo) < 1e-7:
          break

len(datos), media_actual, desviacion_actual
```

```python
# Visualizar la nueva distribución con los datos actuales
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(datos, kde=True)
plt.title("Nueva Distribución con Datos Añadidos")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.show()
```

```python
longitud_original = len(data_buen_conductor)
longitud_datos = len(datos)

# Restaurar data_buen_conductor a su estado original y luego corregir
data_buen_conductor = df[df['material'] == 'Conductor']

# Reemplazar valores en la columna 'u' para los registros existentes
data_buen_conductor = data_buen_conductor.reset_index(drop=True)
data_buen_conductor['u'] = datos[:len(data_buen_conductor)]

# Si hay datos adicionales, añadir nuevas filas con NaN para las demás columnas
if longitud_datos > longitud_original:
    # Crear DataFrame adicional con NaN y valores de 'u' para los datos restantes
    datos_adicionales = pd.DataFrame({
        'w': np.nan,
        'sigma': np.nan,
        'u': datos[longitud_original:],
        'e': np.nan,
        'material': 'Conductor'
    })
    # Añadir este DataFrame a data_buen_conductor
    data_buen_conductor = pd.concat([data_buen_conductor, datos_adicionales], ignore_index=True)

# Verificar con las primeras filas
data_buen_conductor.head(50)
```

```python
from sklearn.linear_model import LinearRegression
# Filtrar los datos para obtener solo las filas completas
data_completa = data_buen_conductor.dropna()

# Separar la variable independiente (X) y la dependiente (y)
X = data_completa[['u']]
y_sigma = data_completa['sigma']
y_w = data_completa['w']
y_e = data_completa['e']

modelo_w = LinearRegression()
modelo_w.fit(X, y_w)

modelo_e = LinearRegression()
modelo_e.fit(X, y_e)

modelo_sigma = LinearRegression()
modelo_sigma.fit(X, y_sigma)

# Mostrar las primeras filas de data_completa para verificar
data_completa.tail()
```

```python
# Filtrar los datos para obtener solo las filas con 'w' faltante
datos_con_w_faltante = data_buen_conductor[data_buen_conductor['w'].isna()]
predicciones_w = modelo_w.predict(datos_con_w_faltante[['u']])
data_buen_conductor.loc[data_buen_conductor['w'].isna(), 'w'] = predicciones_w

datos_con_e_faltante = data_buen_conductor[data_buen_conductor['e'].isna()]
predicciones_e = modelo_e.predict(datos_con_e_faltante[['u']])
data_buen_conductor.loc[data_buen_conductor['e'].isna(), 'e'] = predicciones_e

datos_con_sigma_faltante = data_buen_conductor[data_buen_conductor['sigma'].isna()]
predicciones_sigma = modelo_sigma.predict(datos_con_sigma_faltante[['u']])
data_buen_conductor.loc[data_buen_conductor['sigma'].isna(), 'sigma'] = predicciones_sigma

data_filtrada = data_buen_conductor[data_buen_conductor['sigma'] / (data_buen_conductor['w'] * data_buen_conductor['e']) > 25]
data_buen_conductor = data_filtrada.copy()
data_buen_conductor.tail()
```

```python
# Filtrar los datos para obtener solo aquellos cuyo material es 'Pérdidas'
data_perdidas = df[df['material'] == 'Perdidas']
data_perdidas['u'].values
```

```python
minimo = 1.2566370614359173e-06
maximo = 0.0009424777960769379
media_objetivo = (maximo + minimo) / 2

# La desviación estándar de una distribución uniforme es sqrt((max - min)^2 / 12)
desviacion_objetivo = ((maximo - minimo) ** 2 / 12) ** 0.5

# Datos iniciales
data = np.array([2.50000000e-06, 2.25000000e-06, 2.28000000e-06, 1.25663706e-06,
       1.25663706e-06, 2.51327412e-06, 1.25663706e-06, 2.51327412e-06,
       1.25663706e-06, 1.25663706e-06, 1.25663706e-06, 1.25663706e-06,
       1.25663706e-06, 9.42477796e-04, 1.25663706e-06, 1.25663706e-06,
       1.25663706e-06, 1.25663706e-06, 1.25663706e-06, 1.13097336e-05,
       1.25663706e-06, 1.25663706e-06, 1.25663706e-06])

np.random.seed(0)  # Para reproducibilidad
contador = len(data)  # Contador inicial

while contador < 350:
    nuevo_dato = np.random.normal(media_objetivo, desviacion_objetivo)
    if nuevo_dato >= 0:
        data = np.append(data, nuevo_dato)
        media_actual = np.mean(data)
        desviacion_actual = np.std(data)
        contador += 1

        if abs(media_actual - media_objetivo) < 1e-5 and abs(desviacion_actual - desviacion_objetivo) < 1e-5:
            break

# Resultados finales
len(data), np.mean(data), np.std(data)
```

```python
# Visualizar la nueva distribución con los datos actuales
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(data, kde=True)
plt.title("Nueva Distribución con Datos Añadidos")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Reemplazar valores en la columna 'u' para los registros existentes en data_perdidas
longitud_original_perdidas = len(data_perdidas)
longitud_data = len(data)

data_perdidas = df[df['material'] == 'Perdidas']

data_perdidas = data_perdidas.reset_index(drop=True)
data_perdidas['u'] = data[:longitud_original_perdidas]

# Si hay datos adicionales, añadir nuevas filas con NaN para las demás columnas
if longitud_data > longitud_original_perdidas:
    datos_adicionales_perdidas = pd.DataFrame({
        'w': np.nan,
        'sigma': np.nan,
        'u': data[longitud_original_perdidas:],
        'e': np.nan,
        'material': 'Perdidas'
    })
    data_perdidas = pd.concat([data_perdidas, datos_adicionales_perdidas], ignore_index=True)

# Mostrar las primeras 50 filas para verificar
data_perdidas.head(50)
```

```python
# Filtrar los datos de 'Pérdidas' para obtener solo las filas completas
data_completa_perdidas = data_perdidas.dropna()

# Separar la variable independiente (X) y las variables dependientes (y)
X_perdidas = data_completa_perdidas[['u']]
y_sigma_perdidas = data_completa_perdidas['sigma']
y_w_perdidas = data_completa_perdidas['w']
y_e_perdidas = data_completa_perdidas['e']

# Entrenar los modelos para 'sigma', 'w' y 'e'
modelo_w_perdidas = LinearRegression()
modelo_w_perdidas.fit(X_perdidas, y_w_perdidas)

modelo_e_perdidas = LinearRegression()
modelo_e_perdidas.fit(X_perdidas, y_e_perdidas)

modelo_sigma_perdidas = LinearRegression()
modelo_sigma_perdidas.fit(X_perdidas, y_sigma_perdidas)

# Predecir y rellenar los valores faltantes para 'sigma', 'w' y 'e' en los datos de 'Pérdidas'
datos_con_w_faltante_perdidas = data_perdidas[data_perdidas['w'].isna()]
if not datos_con_w_faltante_perdidas.empty:
    predicciones_w_perdidas = modelo_w_perdidas.predict(datos_con_w_faltante_perdidas[['u']])
    data_perdidas.loc[data_perdidas['w'].isna(), 'w'] = predicciones_w_perdidas

datos_con_e_faltante_perdidas = data_perdidas[data_perdidas['e'].isna()]
if not datos_con_e_faltante_perdidas.empty:
    predicciones_e_perdidas = modelo_e_perdidas.predict(datos_con_e_faltante_perdidas[['u']])
    data_perdidas.loc[data_perdidas['e'].isna(), 'e'] = predicciones_e_perdidas

datos_con_sigma_faltante_perdidas = data_perdidas[data_perdidas['sigma'].isna()]
if not datos_con_sigma_faltante_perdidas.empty:
    predicciones_sigma_perdidas = modelo_sigma_perdidas.predict(datos_con_sigma_faltante_perdidas[['u']])
    data_perdidas.loc[data_perdidas['sigma'].isna(), 'sigma'] = predicciones_sigma_perdidas

# Mostrar las últimas 50 filas para verificar la integración de los datos
data_perdidas.tail(50)
```

```python
# Aplicar el filtro de la condición para 'Pérdidas', donde sigma / (we) debe ser menor que 25
data_filtrada_perdidas = data_perdidas[data_perdidas['sigma'] / (data_perdidas['w'] * data_perdidas['e']) < 25]

# Mostrar las últimas 50 filas para verificar el filtro
data_perdidas = data_filtrada_perdidas.copy()
data_perdidas.tail()
```
